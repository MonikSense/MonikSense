```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "https://raw.githubusercontent.com/MonikSense/MonikSense/refs/heads/main/MXENE%201%20L.csv"
data = pd.read_csv(file_path)

# Set up the figure
fig, axes = plt.subplots(1, 4, figsize=(12, 3))

# Scatter plots
sns.scatterplot(x=data["Time(s)"], y=data["Resistance (Mohm)"],color='red', alpha=0.7, ax=axes[0])
sns.scatterplot(x=data["Electrode Distance (mm)"], y=data["Resistance (Mohm)"], alpha=0.7, ax=axes[1])
sns.scatterplot(x=data["LA Conc (mM)"], y=data["Resistance (Mohm)"], alpha=0.7, ax=axes[2])

# Histogram of Resistance
sns.histplot(data["Resistance (Mohm)"], bins=10, ax=axes[3])

# Labels
axes[0].set_xlabel("Time(s)")
axes[1].set_xlabel("Electrode Distance (mm)")
axes[2].set_xlabel("LA Conc (mM)")
axes[3].set_xlabel("Resistance (Mohm)")
axes[0].set_ylabel("Resistance (Mohm)")

plt.tight_layout()
plt.show()
```
